## Importing Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
house_df = pd.read_csv("housing.csv", delim_whitespace=True, header=None)
house_df.head() #esep='\s+
```

## Setting up the columns name

```python
house_df.columns = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
```

```python
house_df.shape
```

```python
house_df.head()
```

## Checking for missing values

```python
house_df.isna().sum()
```

## Statistical summary of data

```python
house_df.describe()
```

### Understanding the correlation between various features in the dataset using heatmap

```python
plt.figure(figsize=(10,10))
sns.heatmap(house_df.corr(), cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
plt.show()
# cbar - color bar
# fmt = '.1f' - one value after decimal
```

## Splitting the data into features and label

```python
X = house_df.drop(columns='MEDV', axis=1) # 1 for column
Y = house_df['MEDV']
```

```python
X,Y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

## Training the model

```python
model = XGBRegressor()
```

```python
model.fit(X_train, y_train)
```

## Evaluating model

```python
training_data_pred = model.predict(X_train) # predicted value
training_data_pred
```

#### prediction on training data

```python
score = metrics.r2_score(y_train, training_data_pred)
score
```

```python
score_1 = metrics.mean_absolute_error(y_train, training_data_pred)
score_1
```

```python
plt.scatter(y_train, training_data_pred)
plt.xlabel('Actual price')
plt.ylabel('Predicted Price')
plt.title("Actual price Vs Predicted price (Training data)")
plt.show()
```

#### prediction on test data

```python
test_data_pred = model.predict(X_test)
test_data_pred
```

```python
test_score = metrics.r2_score(y_test, test_data_pred)
test_score
```

```python
test_score1 = metrics.mean_absolute_error(y_test, test_data_pred)
test_score1
```

```python
plt.scatter(y_test, test_data_pred)
plt.xlabel('Actual price')
plt.ylabel('Predicted Price')
plt.title("Actual price Vs Predicted price (Test data)")
plt.show()
```
